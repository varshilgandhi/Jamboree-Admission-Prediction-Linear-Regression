```python
import os
import pandas as pd
import functools
cwd = os.getcwd()
```

```python
# Define the directory to start the search
directory = cwd+'/Data/'
```

```python
def first_2chars(x):
    return(x[:2])
```

```python
# Define the directory to start the search
directory = cwd+'/Data/'

type = ['CAT', 'CAT x TAX']
years = ['2024','2023']



# Recursively walk through the directory
for i in type: 
    # List to hold all dataframes
    dataframes = []
    for j in years:
        for root, dirs, files in os.walk(directory+'/'+i+'/'+j+'/'):
            print(directory+'/'+i+'/'+j+'/')
            for file in sorted(files, key = first_2chars):
                if file.endswith('.xlsx'):  # Change to '.csv' for CSV files
                    # Full path to the file
                    file_path = os.path.join(root, file)
                    print(file_path)
                    
                    # Read the file into a DataFrame
                    df = pd.read_excel(file_path,skiprows=range(0, 3))  # Use pd.read_csv() for CSV files
                    
                    # Add the DataFrame to the list
                    dataframes.append(df)
    # Optionally, concatenate all DataFrames into one
    if i == 'CAT':
        if dataframes:
            df_cat = pd.concat(dataframes, ignore_index=True)
    elif i == 'CAT x TAX':
        if dataframes:
            df_cat_tax = pd.concat(dataframes, ignore_index=True)
```

```python
df_qc = pd.concat(df_agg, axis=1, join='outer').fillna(0).reset_index('PLACEMENT NAME')[['PLACEMENT NAME',
    'IMPRESSIONS CAT','IMPRESSIONS FINAL','IMPRESSIONS INITIAL','IMPRESSIONS PLA AUC','IMPRESSIONS SPONSORSHIP',
    'SPEND CAT','SPEND FINAL','SPEND INITIAL','SPEND PLA AUC','SPEND SPONSORSHIP',
    'CLICKS CAT','CLICKS FINAL','CLICKS INITIAL','CLICKS PLA AUC','CLICKS SPONSORSHIP']]

df_qc['IMPRESSIONS TAX QC %'] = (df_qc['IMPRESSIONS FINAL'] - df_qc['IMPRESSIONS INITIAL'])/df_qc['IMPRESSIONS INITIAL']*100
df_qc['IMPRESSIONS CAT - TAX QC %'] = (df_qc['IMPRESSIONS CAT'] - df_qc['IMPRESSIONS FINAL'])/df_qc['IMPRESSIONS FINAL']*100

df_qc['SPEND TAX QC %'] = (df_qc['SPEND FINAL'] - df_qc['SPEND INITIAL'])/df_qc['SPEND INITIAL']*100
df_qc['SPEND CAT - TAX QC %'] = (df_qc['SPEND CAT'] - df_qc['SPEND FINAL'])/df_qc['SPEND FINAL']*100

df_qc['CLICKS TAX QC %'] = (df_qc['CLICKS FINAL'] - df_qc['CLICKS INITIAL'])/df_qc['CLICKS INITIAL']*100
df_qc['CLICKS CAT - TAX QC %'] = (df_qc['CLICKS CAT'] - df_qc['CLICKS FINAL'])/df_qc['CLICKS FINAL']*100

df_qc = df_qc[['PLACEMENT NAME',
    'IMPRESSIONS CAT - TAX QC %','SPEND CAT - TAX QC %','CLICKS CAT - TAX QC %',
    'IMPRESSIONS TAX QC %','SPEND TAX QC %','CLICKS TAX QC %',
    'IMPRESSIONS CAT','IMPRESSIONS FINAL','IMPRESSIONS INITIAL','IMPRESSIONS PLA AUC','IMPRESSIONS SPONSORSHIP',
    'SPEND CAT','SPEND FINAL','SPEND INITIAL','SPEND PLA AUC','SPEND SPONSORSHIP',
    'CLICKS CAT','CLICKS FINAL','CLICKS INITIAL','CLICKS PLA AUC','CLICKS SPONSORSHIP']]

df_qc
```
